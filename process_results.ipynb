```python
df = pd.read_csv("human_study/data/processed/gpt-4o-mini-2024-07-18_mop_new_human_stories_annotations.csv")
df.iloc[:2].T
```

```python
# df has a row for each story and each persona
df["participant_id"].value_counts(), df["story_id"].value_counts()
```

```python
# Average scores across the five personas.
COLS_TO_AGG = [
    'authenticity_score',
    'emotion_provoking_score',
    'empathy_score',
    'engagement_score',
    'narrative_complexity_score',
    'human_likeness_score',
    'time_taken',
]
KEYS = ["story_id"]
```

```python
df_agg = df.groupby(KEYS)[COLS_TO_AGG].mean().reset_index()
df_agg
```
